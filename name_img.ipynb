```python
from bs4 import BeautifulSoup
from splinter import Browser
import requests
import pymongo
import pandas as pd
import pprint
import json
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
# Define the 'classDB' database in Mongo
db = client.projDB
# char_img = db.char_img.find()
```

```python
home_url = 'https://www.hbo.com/'
base_url = 'https://www.hbo.com/game-of-thrones/cast-and-crew'
url = base_url
url
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
search_result_list = []
# results_main = soup.find_all('div', class_='modules/Cast--castGroup')
results_main = soup.find_all('div', class_='modules/Cast--castMember')
results_main
```

```python
character_list = []
for result in results_main:
    main1 = result.find('div', class_='components/ThumbnailWithText--container')
    main2 = main1.find('a')
    name_list = main2["data-bi-interactive"].split("/")
    name = name_list[(len(name_list)-1)]
    main3 = main1.find('img')
    img_url = main3["srcset"].split("/_jcr")[0]
    character_dict = {}
    character_dict["name"] = name.lstrip()
    character_dict["img"] = home_url + img_url
    character_list.append(character_dict)
    db.char_img.insert_one(
                {
                'name': character_dict["name"],
                'img_url': character_dict["img"],
                })

character_list
```
